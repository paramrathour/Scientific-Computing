### Importing Libraries

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

We know that
$$m\ddot{x}=mg-kx$$
Let
$$y=\begin{bmatrix}x&\dot{x}\end{bmatrix}\quad\rightarrow\quad \dot{y}=\begin{bmatrix}\dot{x}&\ddot{x}\end{bmatrix}$$

```python
def model(y,t,k,m,g):
    '''
    function that returns dy/dt
    '''
    v = y[1]                    # dx/dt is velocity
    a = g - y[0]*k/m            # d^2 x/dt^2 is acceleration
    dydt = [v,a]
    return dydt
```

```python
# Initial Conditions and constants (in SI units)
y_0 = [0,0]
k = 2.5
m = 1.5
g = 9.8
```

```python
t = np.linspace(0,10,10000) # Time Points
```

```python
y = odeint(model, y_0, t, args=(k,m,g)) # Solving the System
```

```python
plt.plot(t,y[:,0],'b-',label=r'Position')
plt.plot(t,y[:,1],'r--',label=r'Velocity')
plt.ylabel('Magnitude')
plt.xlabel('Time')
plt.legend(loc='best')
plt.show()
```
